## Example plot

```python
# setup subplots

fig, axs = plt.subplots(15, 15, figsize=(150, 150))
fig.suptitle('Scatterplots of the 16 predictors')

for x in range(0,15):
    for y in range(0, 15):
        # filter if y less than or equal to x 
        if y <= x:
            pass
        else:
            # add plt to subplot
            axs[x, i].scatter(data[0:20, x], data[0:20, y], color = 'red', label = 'Ss')
            axs[x, i].scatter(data[20:40, x], data[20:40, y], color = 'blue', label = 'Sc')
            axs[x, i].scatter(data[40:60, x], data[40:60, y], color = 'green', label = 'Sv')
            axs[x, i].set_xlabel(predictors[x])
            axs[x, i].set_ylabel(predictors[y])
            axs[x, i].legend()

# save plt

plt.savefig('scatterplots.png')

# display scatterplots.png

Image(filename='scatterplots.png')
```

## Generate tensorflow model

```python
# generate tensorflow model from the predictors to predict the species type

# create the model

model = keras.Sequential([
    keras.layers.Dense(16, activation=tf.nn.relu, input_shape=(16,)),
    keras.layers.Dense(3, activation=tf.nn.softmax)
])

# compile the model

model.compile(optimizer='adam',
                loss='sparse_categorical_crossentropy',
                metrics=['accuracy'])
```

```python
# generates labels where the first 20 are Ss, second 20 are Sc and third 20 are Sv

labels = np.zeros(60)
labels[0:20] = 0
labels[20:40] = 1
labels[40:60] = 2

# train the model

model.fit(data, labels, epochs=500)
```

```python
# save model to "Evolution Practical/model.h5"

model.save('model.h5')
```

```python
# evaluate the model

test_loss, test_acc = model.evaluate(data, labels)

print('Test accuracy:', test_acc)

# print model summary

model.summary()
```
